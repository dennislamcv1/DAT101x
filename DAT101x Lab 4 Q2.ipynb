### Import Libraries (Exercise 3: Training a Classification Model)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import math

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

pd.set_option("display.max_rows", 400)
```

### Load data and explore

```python
df = pd.read_csv("lab4q2.csv")
```

### Transform data for training

```python
#Drop Date and Day columns
df = df.drop(["Date"],axis=1)
```

```python
df
```

```python
X_cat = df["Day"]
```

```python
X_cat.head()
```

```python
X_cat
```

```python
# Label encoding for Day feature
le = LabelEncoder()
```

```python
le.fit_transform(X_cat)
```

```python
le.classes_
```

```python
X_encoded = le.fit_transform(X_cat)
```

```python
X_encoded = pd.DataFrame(X_encoded)
```

```python
X_encoded
```

```python
df2 = pd.concat([df,X_encoded],axis=1)
```

```python
df2
```

```python
df2 = df2.drop("Day",axis=1)
```

```python
df2.head()
```

```python
df2.rename(columns={0: "Day"},inplace=True)
```

```python
df2.head()
```

```python
df2.shape
```

```python
#Rearrange columns
df2 = df2[['Sales', 'Temperature', 'Flyers', 'rainfall_log_value', 'Price','Day','Profitable']]
```

```python
df2.head()
```

```python
X = df2.iloc[:,0:6]
y = df2.iloc[:,6]
```

```python
X.head()
```

```python
y.head()
```

```python
X.values, y.values
```

### Model Training

```python
#Split for training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, shuffle=True)
```

```python
X_train.shape, X_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict  #Prediction results
```

```python
y_test.values.reshape(-1,1)
```

### Model Evaluation

```python
accuracy = accuracy_score(y_test,y_predict)
print('Accuracy Score is',accuracy)
```

```python
cm = confusion_matrix(y_test,y_predict)
print(cm)
```

```python
print(classification_report(y_test,y_predict))
```
