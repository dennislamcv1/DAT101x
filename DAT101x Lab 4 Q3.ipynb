Visualize the output of the Customers.csv dataset, and note that it contains observations for 109 customers, including the following features:

• CustomerID: A unique identifier for each customer.

• Name: The customer’s full name.

• Age: The age of the customer.

• AvgWeeklySales. The average number of sales to this customer per week.

• AvgDrinks: The average number of drinks purchased by this customer per sale.

### Transform data for training

```python
# Standardize for Age, AvgWeeklySales, AvgDrinks feature
scaler = StandardScaler()
```

```python
df1 = scaler.fit_transform(df[['Age', 'AvgWeeklySales', 'AvgDrinks']])
```

```python
df1
```

```python
df1 = pd.DataFrame(df1,columns=['Age', 'AvgWeeklySales', 'AvgDrinks'])
```

```python
df1
```

```python
df.head()
```

```python
df = df.drop(['Age', 'AvgWeeklySales', 'AvgDrinks'],axis=1)
```

```python
df.head()
```

```python
df2 = pd.concat([df,df1],axis=1)
```

```python
X = df2.iloc[:,2:5]
```

### Clustering Model Training

```python
model = KMeans(n_clusters=3,random_state=0,max_iter=200)
```

```python
model.fit(X)
```

```python
labels = model.labels_
labels
```

```python
labels = pd.DataFrame(labels,columns=['Cluster'])
```

```python
labels
```

```python
df3 = pd.concat([df2,labels],axis=1)
```
