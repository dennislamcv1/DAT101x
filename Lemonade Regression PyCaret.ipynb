```python
import pandas as pd

from pycaret.regression import *
```

```python
df = pd.read_csv('Lemonade.csv')
```

```python
df
```

```python
df.shape
```

### Setup the Experiment

```python
exp = setup(data=df, target= 'Sales', session_id=0, train_size=0.8, normalize=True, numeric_features=['Price'])
```

```python
compare_models()
```

### Linear Regression Model

```python
lrmodel = create_model('lr')
```

```python
print(lrmodel)
```

```python
plot_model(lrmodel)
```

```python
plot_model(lrmodel, plot = 'error')
```

```python
predict_model(lrmodel)
```

### Predict Test Data

```python
testdata = pd.read_csv("test.csv")
```

```python
testdata
```

```python
prediction = predict_model(estimator=lrmodel, data=testdata, round=2)
```

```python
prediction
```

### Extra Trees Regressor Model

```python
etmodel = create_model(estimator='et')
```

```python
print(etmodel)
```

### Hyperparameter Tuning

```python
tuned_et = tune_model('et', n_iter=25)
```

```python
print(tuned_et)
```

### Model Plot

```python
plot_model(tuned_et)
```

```python
plot_model(tuned_et, plot = 'error')
```

```python
plot_model(tuned_et, plot='feature')
```

### Model Evaluation

```python
evaluate_model(tuned_et)
```

### Model Prediction

```python
predict_model(tuned_et)
```

### Finalize Model

```python
final_et = finalize_model(tuned_et)
```

```python
print(final_et)
```

```python
predict_model(final_et)
```

```python
testdata = pd.read_csv("test.csv")
```

```python
testdata
```
